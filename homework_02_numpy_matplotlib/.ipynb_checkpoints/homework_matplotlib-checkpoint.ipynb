```python
%matplotlib inline
import pylab
import numpy as np
from matplotlib.pyplot import *
from ipywidgets import interact #intersting
def Quadratic(x): # å®šä¹‰äºŒæ¬¡å‡½æ•°
    return 2*x**2 +3*x +4
def plotLadder(laddernum):
    x = np.linspace(-5, 5, num=100)
    y = Quadratic(x)
    plot(x,y) # å…ˆç”»å‡ºåŽŸå‡½æ•°çš„å›¾å½¢
    
    a = np.linspace(-5, 5, num=laddernum)
    for i in range(laddernum):
        plot([a[i],a[i]],[0,Quadratic(a[i])],color="black") # ç”»æ¢¯å½¢çš„ä¸Šåº•å’Œä¸‹åº•

    ladders = [];
    for i in range(laddernum):
        ladders.append([a[i],Quadratic(a[i])]) # å› ä¸ºæ¢¯å½¢çš„è…°æ˜¯å‘ˆä¸€æ¡ç›´çº¿ï¼Œæ‰€ä»¥è¿™é‡Œå­˜ä¸‹å„ç‚¹åæ ‡
    
    npladders = np.array(ladders)
    plot(npladders[:,0],npladders[:,1]); # æŠŠæ¢¯å½¢çš„æ–œè…°è¿žèµ·æ¥
def plotSinFun(x):
    x = np.linspace(x[0],x[1],num = 30)
    y = np.power(np.sin(x - 2),2)*np.exp(-np.power(x,2))
    figure(1)
    plot(x,y)
    title(r"$sin^2(x - 2) e^{-x^2}$")
    xlabel("x label")
    ylabel("y label")

    

#homework_matplotlib_01
interact(plotLadder, laddernum=(1, 30, 1)) # æ»‘åŠ¨æ¨¡å—åœ¨ 1 å’Œ 30 ä¹‹é—´å˜åŒ–ï¼Œå˜åŒ–åŒºé—´æ˜¯ 1

#homework_matplotlib_02
x=[0,2]
plotSinFun(x)
#homework_matplotlib_03
nsteps = 1000
draws = np.random.randint(-1,2,size=(2,nsteps))
walks = draws.cumsum(1)
figure(2)
plot(walks[0,:],walks[1,:]);
#homework_matplotlib_04
#homework_matplotlib_05
#homework_matplotlib_06
```
